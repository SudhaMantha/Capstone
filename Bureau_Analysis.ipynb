## Number of Loans per customer

```python
grp = bur[['SK_ID_CURR', 'DAYS_CREDIT']].groupby(by = ['SK_ID_CURR'])['DAYS_CREDIT'].count().reset_index().rename(index=str, columns={'DAYS_CREDIT': 'BUREAU_LOAN_COUNT'})
```

```python
grp[grp['SK_ID_CURR'] == 215354]
```

```python
bur = bur.merge(grp, on = ['SK_ID_CURR'], how = 'left')
```

## Types of loan ( may be not needed so not merging)

## Days_Credit_Enddate if -ve mean red flag (past)

```python
bur['CREDIT_ENDDATE_BINARY'] = bur['DAYS_CREDIT_ENDDATE']

def f(x):
    if x<0:
        y = 0
    else:
        y = 1   
    return y

bur['CREDIT_ENDDATE_BINARY'] = bur.apply(lambda x: f(x.DAYS_CREDIT_ENDDATE), axis = 1)

```

```python
grp = bur.groupby(by = ['SK_ID_CURR'])['CREDIT_ENDDATE_BINARY'].mean().reset_index().rename(index=str, columns={'CREDIT_ENDDATE_BINARY': 'CREDIT_ENDDATE_PERCENTAGE'})
```

```python
grp[grp['SK_ID_CURR'] == 215354]
```

```python
bur = bur.merge(grp, on = ['SK_ID_CURR'], how = 'left')
```
